```python
import pandas as pd
df=pd.read_csv('FAOSTAT_data_en_12-29-2024.csv')
df.head()
df.shape
```

```python
columns_to_drop = ['Domain Code', 'Domain', 'Area Code (M49)', 'Area', 'Element','Year Code','Unit','Year','Note']
df = df.drop(columns=columns_to_drop)
df.head()
```

```python
df.columns=[x.lower() for x in df.columns]
```

```python
df['recentered']=df['value']-df['value'].mean()
df.head()
```

```python
fresh_produce=df['item'].str.contains('fresh')
df.loc['recentered']=df.loc[fresh_produce,'value']-df.loc[fresh_produce].value.mean()
df[fresh_produce]
```

```python
def remove_space(s):
  result=''
  for i in s:
    if i==' ':
      continue
    result+=i
  return result


df['item']=df['item'].apply(remove_space)
df.head()
```
